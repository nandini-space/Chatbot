```python
print("Luca: Hello! I am Luca, your friendly AI chatbot.")

name = input("Before we start, what's your name? ")

print(f"Welcome, {name}!")
print("Type 'bye' to exit.\n")

responses = {
    "hello": f"Hello {name}! How can I help you?",
    "hi": f"Hi {name}!",
    "hey": f"Hey {name}! Nice to see you.",
    "how are you": "I'm doing great! Thanks for asking.",
    "what is your name": "My name is Luca.",
    "who created you": "I was created using Python.",
    "thanks": "You're welcome!",
    "thank you": "Happy to help!",
    "joke": "Why do programmers prefer dark mode? Because light attracts bugs!",
    "ai": "Artificial Intelligence helps machines think and learn.",
    "python": "Python is one of the most popular programming languages."
}

while True:
    user_input = input(f"{name}: ").lower().strip()

    if user_input in ["bye", "exit", "quit"]:
        print("🤖 Luca: Goodbye! It was great chatting with you. Have an amazing day!")
        break

    elif "college" in user_input:
        print("🤖 Luca: That's awesome! What course are you pursuing?")

    elif "age" in user_input:
        print("🤖 Luca: I'm timeless! 😄 How old are you?")

    elif "favorite color" in user_input:
        print("🤖 Luca: I love blue 💙 because it reminds me of technology and innovation!")

    elif "study" in user_input:
        print("🤖 Luca: Keep learning and growing. Every skill you learn opens new opportunities!")

    elif "goal" in user_input:
        print("🤖 Luca: Having clear goals is the first step toward success. What's your goal?")

    else:
        print(
            responses.get(
                user_input,
                "🤖 Luca: Hmm, I'm not sure about that yet. Could you ask me something else?"
            )
        )
```
